```python
d3 = pd.DataFrame({'Roll No,': [1,2,3,4,5,6,7,9],'Physics':[32,53,45,43,34,23,45,67]})
d4 = pd.DataFrame({'Roll No,': [1,2,3,4,5,8,11,12],'Chemistry':[32,53,45,43,34,65,34,87]})
```

```python
pd.merge(d3,d4, how = 'right')
```

```python
pd.merge(d3,d4, how = 'left')
```

```python
pd.merge(d3,d4)
```

```python
pd.merge(d3,d4, how = 'outer')
```
